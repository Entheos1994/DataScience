```python
ingredients = []
cuisines = []
for el in train:
    for ingr in el['ingredients']:
        if ingr not in ingredients:
            ingredients.append(ingr)
    if el['cuisine'] not in cuisines:
        cuisines.append(el['cuisine'])
ingredients
cuisines
```

```python
dict_ingredients = {}
for ingr in ingredients:
    dict_ingredients[ingr]=0
```

```python
dict_cuisines = {}
for cuis in cuisines:
    dict_cuisines[cuis]=0
dict_cuisines
```

```python
mapping = {}
mapping.clear()
for ingr in ingredients:
    mapping[ingr] = dict_cuisines.copy()

mapping
```

```python
for el in train:
    c = el['cuisine']
    for ingr in el['ingredients']:
        mapping[ingr][c] += 1
mapping
```

```python
for el in mapping:
    count = 0
    for cuisine in mapping[el]:
        count += mapping[el][cuisine]
    for cuisine in mapping[el]:
        mapping[el][cuisine] = mapping[el][cuisine]/float(count)
```
